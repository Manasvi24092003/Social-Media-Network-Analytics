Problem statement 2: 
There are three datasets given (Facebook, Instagram, and LinkedIn). Construct and visualize the following networks:
●	circular network for Facebook
●	star network for Instagram
●	star network for LinkedIn

Create a network using an adjacency matrix (undirected only). The snapshots of those datasets are given below: 


```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

```

```python

# Load the datasets (Facebook, Instagram, LinkedIn)
facebook_df = pd.read_csv(r"C:\Users\manas\Downloads\Data Science 360Dgtmg\Network Analytics\Datasets\facebook.csv" )   # Adjacency matrix for Facebook
instagram_df = pd.read_csv(r"C:\Users\manas\Downloads\Data Science 360Dgtmg\Network Analytics\Datasets\instagram.csv") # Adjacency matrix for Instagram
linkedin_df = pd.read_csv(r"C:\Users\manas\Downloads\Data Science 360Dgtmg\Network Analytics\Datasets\linkedin.csv" )
```

# Circular Network for Facebook

```python
print(facebook_df.index)
print(facebook_df.columns)
```

```python
# Ensure columns and index match
facebook_df.columns = facebook_df.index
```

```python
# Create an undirected graph from the adjacency matrix
facebook_graph = nx.from_pandas_adjacency(facebook_df)

# Visualize the circular network
plt.figure(figsize=(8, 8))
pos = nx.circular_layout(facebook_graph)
nx.draw(facebook_graph, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')
plt.title("Facebook Circular Network")
plt.show()
```

# Star Network for Instagram

```python
#Instagram: Star Network
print(instagram_df.index)
print(instagram_df.columns)
```

```python
# Ensure columns and index match
instagram_df.columns = instagram_df.index
```

```python
central_node = max(instagram_graph.degree, key=lambda x: x[1])[0]

# Provide initial positions for the fixed node (central node)
initial_pos = {central_node: (0, 0)}  # Fix the central node at the origin (0, 0)

# Use spring_layout with the central node fixed
pos = nx.spring_layout(instagram_graph, pos=initial_pos, fixed=[central_node])

# Visualize the star network
plt.figure(figsize=(2, 2))
nx.draw(instagram_graph, pos, with_labels=True, node_color='lightgreen', node_size=500, edge_color='gray')
plt.title("Instagram Star Network")
plt.show()
```

# Star Network for LinkedIn

```python
#Instagram: Star Network
print(linkedin_df.index)
print(linkedin_df.columns)
```

```python
# Ensure columns and index match
linkedin_df.columns = linkedin_df.index
```

```python
# Find the most connected node (central node) for the star layout
central_node_linkedin = max(linkedin_graph.degree, key=lambda x: x[1])[0]

# Initialize positions with the central node at the center
initial_pos = {central_node_linkedin: (0, 0)}

# Generate positions for the rest of the nodes using spring layout
# This will place the central node at (0, 0) and compute positions for other nodes
pos = nx.spring_layout(linkedin_graph, pos=initial_pos, fixed=[central_node_linkedin])

# Visualize the star network
plt.figure(figsize=(8, 8))
nx.draw(linkedin_graph, pos, with_labels=True, node_color='orange', node_size=500, edge_color='gray')
plt.title("LinkedIn Star Network")
plt.show()
```
